```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("pfe_with_macro_uncorrelated.csv")
print(df.head())
df['Date'] = pd.to_datetime(df['Date'])  
train_end = pd.to_datetime("2021-12-31")
train_df = df[df['Date'] <= train_end].copy()
test_df = df[df['Date'] > train_end].copy()

train_date = train_df['Date']
test_date = test_df['Date']
train_target = train_df['target_return']
test_target = test_df['target_return']

train_df = train_df.drop(columns=['Date', 'target_return'])
test_df = test_df.drop(columns=['Date', 'target_return'])
train_mean = train_df.mean()
train_std = train_df.std()

train_df = (train_df - train_mean) / train_std
test_df = (test_df - train_mean) / train_std  

train_df['Date'] = train_date.values
train_df['target_return'] = train_target.values

test_df['Date'] = test_date.values
test_df['target_return'] = test_target.values


train_df.set_index('Date', inplace=True)
test_df.set_index('Date', inplace=True)

train_df.to_csv("final_train_normalized.csv", index=True)
test_df.to_csv("final_test_normalized.csv", index=True)
print(train_df.head())
```
